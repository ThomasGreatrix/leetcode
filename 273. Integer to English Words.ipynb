```python
def numberToWords(num):
    """
    Takes in an integer and returns a string
    stating that integer in words.
    """
    # Handles num == 0
    if not num:
        return "Zero"
    
    #####
    # Breakdown the input number into units,
    # tens, hundreds, etc...
    #####
    
    # Units we initially care about.
    units = ['Billion ',
             'Million ',
             'Thousand ',
             'Hundred ',
             'One ']
    
    # The values that we will check for.
    vals = [1000000000, 1000000, 1000, 100, 1]
    
    # Create a dictionary for the amount
    # of each unit num has.
    breakdown = {units[i]: 0 for i in range(len(units))}
    for i in range(5):
        while num >= vals[i]:
            breakdown[list(breakdown.keys())[i]]=breakdown[list(breakdown.keys())[i]]+1
            num = num - vals[i]
    
    #####
    # Create the english way of explaining
    # how much of each unit we have.
    #####
    
    # Series of lists for english words.
    tens = ["Twenty ",
            "Thirty ",
            "Forty ",
            "Fifty ",
            "Sixty ",
            "Seventy ",
            "Eighty ",
            "Ninety "]
    digits = ["One ",
              "Two ",
              "Three ",
              "Four ",
              "Five ",
              "Six ",
              "Seven ",
              "Eight ",
              "Nine ",
              "Ten ",
              "Eleven ",
              "Twelve ",
              "Thirteen ",
              "Fourteen ",
              "Fifteen ",
              "Sixteen ",
              "Seventeen ",
              "Eighteen ",
              "Nineteen "]
    
    # Put the english amounts in the dictionary.
    for i in list(breakdown.keys()):
        # If n < 20, we already have done the
        # work.
        if breakdown[i] < 20 and breakdown[i]:
            breakdown[i] = digits[breakdown[i]-1]
        # Less than 100 requires some more work.
        elif breakdown[i] < 100 and breakdown[i]:
            ten = 0
            while breakdown[i]>=10:
                ten = ten+1
                breakdown[i]=breakdown[i]-10
            if breakdown[i]:
                breakdown[i] = tens[ten-2]+digits[breakdown[i]-1]
            else:
                breakdown[i] = tens[ten-2]
        # More than 100 requires alot more work.
        elif breakdown[i]>=100:
            hundred = 0
            while breakdown[i]>=100:
                hundred = hundred+1
                breakdown[i]=breakdown[i]-100
            ten=0
            if breakdown[i]>=20:
                while breakdown[i]>=10:
                    ten = ten+1
                    breakdown[i]=breakdown[i]-10
            if ten:
                if breakdown[i]:
                    breakdown[i]=digits[hundred-1]+"Hundred "+tens[ten-2]+digits[breakdown[i]-1]
                else:
                    breakdown[i]=digits[hundred-1]+"Hundred "+tens[ten-2]
            else:
                if breakdown[i]:
                    breakdown[i]=digits[hundred-1]+"Hundred "+digits[breakdown[i]-1]
                else:
                    breakdown[i]=digits[hundred-1]+"Hundred "
    
    #####
    # Use the dictionary to construct
    # the final string.
    #####
    
    string = ""
    for i in list(breakdown.keys()):
        if breakdown[i]:
            if i !="One ":
                string = string+breakdown[i]+i
            else:
                string = string+breakdown[i]
    return string
```
